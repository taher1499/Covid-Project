```python
# taking only dates upto the end of the particular week

#required_data = df[df.index <= 256]
#required_data = df[df.index <= 883]
#required_data = df[df.index <= 3105]
required_data = df[df.index <= 7599]
```

```python
# Create a data frame containing data of only hospitalized people 

df_with_only_hospitalized = required_data[required_data['current_status'] == 'Hospitalized']
```

```python
# Removing data with missing details about districts 

df_hospitalized_no_na = df_with_only_hospitalized[df_with_only_hospitalized['detected_district'].notna()]
```

```python
# Create object with number of cases per district

no_of_cases_by_district = df_hospitalized_no_na.groupby(['detected_district']).size()
```

```python
# change object to data frame 

no_of_cases_by_district = no_of_cases_by_district.to_frame()
```

```python
no_of_cases_by_district
```

```python
# rename index column 

no_of_cases_by_district.rename({0 : 'no_of_cases'}, axis = 1, inplace=True)
```

```python
# creating new data frame that contains only districts that are 'hotspots'

hotspots = no_of_cases_by_district[no_of_cases_by_district['no_of_cases'] >= 10]
```

```python
hotspots
```

```python
no_of_hotspots=pd.merge(hotspots,df_hospitalized_no_na,how='left',on='detected_district')
```

```python
number_1=pd.read_csv('number_1.csv')
number_2=pd.read_csv('number_2.csv')
number_3=pd.read_csv('number_3.csv')
number_4=pd.read_csv('number_4.csv')
```

```python
number_1.drop('Unnamed: 0',axis=1,inplace=True)
```

```python
number_2.drop('Unnamed: 0',axis=1,inplace=True)
```

```python
number_3.drop('Unnamed: 0',axis=1,inplace=True)
```

```python
number_4.drop('Unnamed: 0',axis=1,inplace=True)
```

```python
number_4.rename({'State' : 'State_2'}, axis = 1, inplace=True)
```

```python
result = pd.concat([number_1,number_2,number_3,number_4],axis=1)
```

```python
result.drop('State',axis=1,inplace=True)
```

```python
result.fillna(0,inplace=True)
```

```python
result = result[["State_2", "number_of_hotspots_20/3/2020", "number_of_hotspots_27/3/2020","number_of_hotspots_03/4/2020","number_of_hotspots_10/4/2020"]]
```

```python
result
```

```python
result['change1']=result['number_of_hotspots_27/3/2020']-result['number_of_hotspots_20/3/2020']
result['change2']=result['number_of_hotspots_03/4/2020']-result['number_of_hotspots_27/3/2020']
result['change3']=result['number_of_hotspots_10/4/2020']-result['number_of_hotspots_03/4/2020']
```

```python
table1=result[['State_2',"number_of_hotspots_20/3/2020","number_of_hotspots_27/3/2020",'change1']]
table2=result[['State_2',"number_of_hotspots_27/3/2020","number_of_hotspots_03/4/2020",'change2']]
table3=result[['State_2',"number_of_hotspots_03/4/2020","number_of_hotspots_10/4/2020",'change3']]
```

```python
table1.sort_values(by='change1',ascending=False,inplace=True)
```

```python
table2.sort_values(by='change2',ascending=False,inplace=True)
```

```python
table3.sort_values(by='change3',ascending=False,inplace=True)
```

```python
table1.to_csv('change_first_week.csv')
table2.to_csv('change_second_week.csv')
table3.to_csv('change_third_week.csv')
```
